```python
import pandas as pd

# Replace 'file_path.csv' with the path to your CSV file
file_path = '/content/drive/MyDrive/true_output.csv'

# Load the CSV file into a DataFrame
true_output = pd.read_csv(file_path)
true_output.head(30)
```

```python
model_output['Prediction'] = model_output['Prediction'].str.extract(r'Answer\s*:?\s*(?:Option\s*)?(\d+)')
model_output['Prediction'] = model_output['Prediction'].fillna(0).astype(int)  # Replace NaN with 0
for i in range(len(model_output['Prediction'])):
  if model_output['Prediction'][i]==0:
    model_output.drop(i,inplace=True)
    true_output.drop(i,inplace=True)
```

```python
correct_predictions = (true_output['Answer Index'] == model_output['Prediction']).sum()

# Calculate accuracy
total_samples = len(true_output)
accuracy = correct_predictions / total_samples

print(f"Accuracy: {accuracy * 100:.2f}%")
```
