```python
import pandas as pd

# Replace 'file_path.csv' with the path to your CSV file
file_path = '/content/drive/MyDrive/final_evaluation_results.csv'

# Load the CSV file into a DataFrame
model_output = pd.read_csv(file_path)
model_output
```

```python
model_output['Predicted Output'] = model_output['Predicted Output'].str.extract(r'Answer\s*:? :?\s*(?:Option\s*)?(\d+)')
model_output['Predicted Output'] = model_output['Predicted Output'].fillna(0).astype(int)  # Replace NaN with 0
for i in range(len(model_output['Predicted Output'])):
  if model_output['Predicted Output'][i]==0:
    model_output.drop(i, inplace=True)

model_output
```

```python
correct_predictions = (model_output['True Label'] == model_output['Predicted Output']).sum()

# Calculate accuracy
total_samples = len(model_output)
accuracy = correct_predictions / total_samples

print(f"Accuracy: {accuracy * 100:.2f}%")
```
