```python
!pip install transformers datasets torch

from transformers import AutoModelForCausalLM, AutoTokenizer
from datasets import load_dataset
import torch
import re
from tqdm import tqdm

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

model_name = "EleutherAI/gpt-neo-2.7B"
model = AutoModelForCausalLM.from_pretrained(model_name).to(device)
tokenizer = AutoTokenizer.from_pretrained(model_name)

dataset = load_dataset("GBaker/MedQA-USMLE-4-options", split='test')

# Evaluate model on the dataset
correct_predictions = 0
total_predictions = 0
idx = 0
check = 0

def get_model_prediction(question, options_dict, correct_option):
    """
    Generate a zero-shot prediction using GPT-Neo for a question with multiple choices.
    The options are provided as a dictionary with keys: 'A', 'B', 'C', 'D'.
    """
    options_text = ', '.join([f"{key}: {value}" for key, value in options_dict.items()])
    input_text = f"Question: {question}\nOptions: {options_text}\nAnswer:"

    # Tokenize input text
    inputs = tokenizer(input_text, return_tensors="pt").to(device)

    # Generate output (can be tuned using temperature, max_length, etc.)
    output = model.generate(inputs["input_ids"], max_new_tokens=172)

    # Decode the output to text
    generated_text = tokenizer.decode(output[0], skip_special_tokens=True)

    # Extract the answer after "Answer:" using regex
    match = re.search(r"Answer:\s*([A-D])", generated_text)
    
    if match:
        predicted_option = match.group(1)
    else:
        predicted_option = None  # Handle case where no match is found
    
    global idx
    global check
    
    if predicted_option == correct_option:
        check += 1
    
    idx += 1
    
    print(check , "/" , idx)
    
    return predicted_option


for sample in tqdm(dataset):
    question = sample['question']
    options = sample['options']  # This is a dictionary with 'A', 'B', 'C', 'D' as keys
    correct_answer = sample['answer_idx']  # Correct answer is the key ('A', 'B', 'C', or 'D')

    # Get the model's prediction
    predicted_answer = get_model_prediction(question, options, correct_answer)

    # Check if prediction is correct
    if predicted_answer == correct_answer:
        correct_predictions += 1
    
    total_predictions += 1

# Calculate accuracy
accuracy = correct_predictions / total_predictions * 100
print(f"Accuracy: {accuracy:.2f}%")
```
